```python
import spotipy
import json
import webbrowser
import pandas as pd
import urllib.request
import spotipy.util as util
```

```python
url="https://api.open-meteo.com/v1/forecast?latitude=52.52&longitude=13.41&hourly=temperature_2m"
```

```python
request = urllib.request.Request(url)
response = urllib.request.urlopen(request)
```

```python
weather = json.loads(response.read())
print(weather)
```

```python
forecast = weather["hourly"]["temperature_2m"]
forecast
```

```python
temp=forecast[0]
temp_written = ""
if temp >= 25 and temp <=45:
    temp_written="hot"
elif temp<25:
    temp_written = "warm"
```

```python
credentials = "spotify_keys.json"
with open(credentials, "r") as keys:
    api_tokens = json.load(keys)
api_tokens
```

```python
client_id = api_tokens["client_id"]
client_secret = api_tokens["client_secret"]
redirectURI = api_tokens["redirect"]
username = api_tokens["username"]
```

```python
redirectURI
```

```python
scope = 'user-read-private user-read-playback-state user-modify-playback-state playlist-modify-public'
token = util.prompt_for_user_token(username, scope, client_id=client_id,
                           client_secret=client_secret,
                           redirect_uri=redirectURI)
```

```python
sp = spotipy.Spotify(auth=token)
```

```python
devices = sp.devices()
print(devices)
deviceID = devices['devices'][0]['id']
```

```python
track_results = sp.search(q=forecast, type='track', limit=50)
tracks = track_results['tracks']['items']
tracks[0]
```

```python
# URI = uniform resource identifier
track_selection_list = []
track_art = []

for song in tracks:
    track_selection_list.append(song['uri'])
    track_art.append(song['album']['images'][0]['url'])
```

```python
my_playlist = sp.user_playlist_create(user="31mvdqp55nskpk66v4nzpyea42ci", name="forecast", public=True,
                                      description="Songs for the weather")
results = sp.user_playlist_add_tracks(username, my_playlist['id'], track_selection_list)
print(results)
print(my_playlist)
```
